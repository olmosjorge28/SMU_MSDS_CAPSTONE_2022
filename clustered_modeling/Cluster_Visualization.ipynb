```python
import pandas as pd
import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cluster_df = pd.read_csv('Cluster_Plot_Data.csv')
```

```python
cluster_df.head()
```

```python
p = sns.FacetGrid(data=cluster_df, row='Clusters Combined', col='Train_Test', 
            sharey=False, margin_titles=True)
p.map_dataframe(sns.lineplot, x='Timestep', y='Mean_max_diameter', 
                hue='blacklisted')
p.set_axis_labels("Timestep (30 Seconds)", "Average Max_Diameter")
p.set_titles(col_template="{col_name} Set", row_template="Cluster #{row_name}")
p.add_legend(title='Blacklisted')
p.axes[0,0].set_ylim(0,0.6)
p.axes[0,1].set_ylim(0,0.6)
p.axes[1,0].set_ylim(0,2.5)
p.axes[1,1].set_ylim(0,2.5)
p.axes[2,0].set_ylim(0,11)
p.axes[2,1].set_ylim(0,11)
plt.show()
```

```python
p = sns.FacetGrid(data=cluster_df, row='Train_Test', col='Clusters Combined', 
            sharey=False, margin_titles=True)
p.map_dataframe(sns.lineplot, x='Timestep', y='Mean_max_diameter', 
                hue='blacklisted')
p.set_axis_labels("Timestep (30 Seconds)", "Average Max_Diameter")
p.set_titles(col_template="Cluster #{col_name}", row_template="{row_name}")
p.add_legend(title='Blacklisted')
p.axes[0,0].set_ylim(0,0.6)
p.axes[1,0].set_ylim(0,0.6)
p.axes[0,1].set_ylim(0,2.5)
p.axes[1,1].set_ylim(0,2.5)
p.axes[0,2].set_ylim(0,11)
p.axes[1,2].set_ylim(0,11)
plt.show()
```

```python
p = sns.FacetGrid(data=cluster_df[cluster_df['Train_Test']=='Train'], col='Clusters Combined', 
            sharey=False, margin_titles=True)
p.map_dataframe(sns.lineplot, x='Timestep', y='Mean_max_diameter', 
                hue='blacklisted', ci=None)
p.set_axis_labels("Timestep (30 Seconds)", "Average Max_Diameter")
p.set_titles(col_template="Cluster #{col_name}")
p.add_legend(title='Blacklisted')
p.axes[0,0].set_ylim(0,0.6)
p.axes[0,1].set_ylim(0,2.5)
p.axes[0,2].set_ylim(0,11)
plt.show()
```
